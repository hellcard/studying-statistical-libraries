# colors and styles

```python
fig, ax = plt.subplots()

x = np.array([ -3, -2, -1, 0.1, 1, 2, 3 ])
y = np.array([ 9, 4, 1, 0.1, 1, 4, 9 ])

ax.plot(x, y)
ax.plot(x, np.sin(x), color = ( 1.0, 0.2, 0.3 ), linestyle = '-')
ax.plot(x, x+5, color = 'blue', linestyle = '--')
ax.plot(x, x+3, color = 'g', linestyle = ':')
ax.plot(x, np.cos(x), color = '0.75', linestyle = '-.')
ax.plot(x, x, color = '#FFDD44', linestyle = '--')
```

```python
fig, ax = plt.subplots()

x = np.array([ -3, -2, -1, 0.1, 1, 2, 3 ])
y = np.array([ 9, 4, 1, 0.1, 1, 4, 9 ])

ax.plot(x, y, marker = 'o')
ax.plot(0.5 * x, y, marker = '*')
ax.plot(2 * x, y, marker = '^')
```

```python
fig, ax = plt.subplots()

x = np.array([ -3, -2, -1, 0.1, 1, 2, 3 ])
y = np.array([ 9, 4, 1, 0.1, 1, 4, 9 ])

ax.plot(x, y, color = 'red', marker = 'o', linestyle = '--', alpha = 0.3)
```

```python
fig, ax = plt.subplots()

x = np.array([ -3, -2, -1, 0.1, 1, 2, 3 ])
y = np.array([ 9, 4, 1, 0.1, 1, 4, 9 ])

ax.plot(x, y, color = 'red', marker = 'o', linestyle = '--', alpha = 0.9)
```

# signatures and marks

```python
fig, ax = plt.subplots()

x = np.array([ -3, -2, -1, 0.1, 1, 2, 3 ])
y = np.array([ 9, 4, 1, 0.1, 1, 4, 9 ])

ax.plot(x, y, color = 'red', linestyle = '-', alpha = 0.7)
ax.plot(3 * x, y, 'o--g', alpha = 0.8)
```

```python
fig, ax = plt.subplots()

x = np.array([ -3, -2, -1, 0.1, 1, 2, 3 ])
y = np.array([ 9, 4, 1, 0.1, 1, 4, 9 ])

ax.plot(x, y, color = 'red', linestyle = '-', alpha = 0.7, label = 'y = x^2')
ax.plot(3 * x, y, 'o--g', alpha = 0.8, label = 'y  = 3 * x^2')

ax.set_title('TITLE', fontsize = 15)

ax.set_xlabel('X')
ax.set_ylabel('Y')

ax.legend(loc = 'upper left') # upper / lower | left / right
```

```python
# plt.cla() - clear the chart
# plt.clf() - clear the drawing

fig, ax = plt.subplots()

x = np.array([ -3, -2, -1, 0.1, 1, 2, 3 ])
y = np.array([ 9, 4, 1, 0.1, 1, 4, 9 ])

ax.plot(x, y, '*--r', label = 'y = x^2')
```

```python
fig, ax = plt.subplots()

x = np.array([ -3, -2, -1, 0.1, 1, 2, 3 ])
y = np.array([ 9, 4, 1, 0.1, 1, 4, 9 ])

ax.plot(x, y, '*--r', label = 'y = x^2')

ax.set_xticks(np.arange(-3, 3, 0.5))
ax.set_yticks(np.arange(0, 10, 3))
```

```python
fig, ax = plt.subplots()

x = np.array([ -3, -2, -1, 0.1, 1, 2, 3 ])
y = np.array([ 9, 4, 1, 0.1, 1, 4, 9 ])

ax.plot(x, y, '*--r', label = 'y = x^2')

ax.set_xticks(np.arange(-3, 3, 0.5))
ax.set_yticks(np.arange(0, 10, 3))

ax.spines['right'].set_visible(False)
```

```python
fig, ax = plt.subplots()

x = np.array([ -3, -2, -1, 0.1, 1, 2, 3 ])
y = np.array([ 9, 4, 1, 0.1, 1, 4, 9 ])

ax.plot(x, y, '*--r', label = 'y = x^2')

ax.set_xticks(np.arange(-3, 3, 0.5))
ax.set_yticks(np.arange(0, 10, 3))

ax.spines['right'].set_visible(False)

ax.grid(color = 'blue', lw = 0.5, ls = '--', ) # lw = linewidth, ls = linestyle
```

# types of charts

```python
fig, ax = plt.subplots()

n = 50

x = np.random.rand(n)
y = np.random.rand(n)

colors = np.random.rand(n)
area = (30 * np.random.rand(n)) ** 2

ax.scatter(x, y, s = area, c = colors, alpha = 0.6, marker = 'o')
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize = ( 10, 6 ))
ax = fig.add_subplot(111, projection = '3d')

ax.scatter(x, y, s = area, c = colors, alpha = 0.5, marker = 'o')
```

```python
fig, axs = plt.subplots(1,2)
```

```python
axs[0].bar([1, 2, 3], [3, 4, 5], color = 'red', label = 'vertical bar', alpha = 0.7)
axs[1].barh([ 0.5, 1, 2.5 ], [ 0, 1, 2 ])

axs[0].legend()
fig
```

```python
fig = plt.figure(figsize = (14,10))

rect1 = plt.bar(np.arange(5), np.arange(5)**2, width = 0.5, color = 'lightblue')
rect2 = plt.bar(np.arange(5), np.arange(5)*2, width = 0.5, color = 'blue')

plt.legend((rect1[0], rect2[0]), ('1','2'))
```

```python
fig, ax = plt.subplots()

mu, sigma  = 100, 15
x = mu + sigma * np.random.randn(1000)

ax.hist(x, 50, density = True, facecolor = 'g', alpha = 0.75)
```

```python
user1 = [10,15,3,12,13,14]
user2 = [20,12,10,11,17]
data = [user1, user2]
fig = plt.figure(figsize = (8, 6))

ax = fig.add_axes([ 0, 0, 1, 1 ])

bp = ax.boxplot(data)

plt.xticks([1,2],['user-1','user-2'])
```

```python
fig, ax = plt.subplots()

x = np.arange(0,100)
y = x * 2

ax.fill_between(x, y, color = 'skyblue', alpha = 0.2)
ax.plot(x, y, color = 'Slateblue', alpha = 0.6)
```

```python
fig, ax = plt.subplots()

labels = ['September','October','November','December']
user1 = [18,28,13,12]

ax.pie(user1, labels = labels)
```
